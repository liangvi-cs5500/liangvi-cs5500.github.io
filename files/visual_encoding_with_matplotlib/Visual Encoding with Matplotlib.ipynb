```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pkmn = pd.read_csv('Pokemon.csv')
pkmn.head()
```

### Visualization 1

```python
bins=range(100,650, 50) 
plt.hist(pkmn["Total"],bins,histtype="bar")
plt.xlabel('Total Stats')
plt.ylabel('Count')
plt.plot()
plt.axvline(pkmn['Total'].mean(),linestyle='dashed',color='red')
plt.axvline(pkmn['Total'].quantile(0.75),linestyle='dashed',color='red')
plt.axvline(pkmn['Total'].quantile(0.25),linestyle='dashed',color='red')
plt.show()
```

Pokemon comes in a variety of stats and often those with higher stats will do better in battles. In tournment formats, Pokemon are placed in certain tier formats, like weight limits for fighters, based on their usage and total stats. In this visualization, we use a histogram to plot the continous field of total stats with the number of pokemon that falls in that bin. The horizontal space convey the progress of stats, while the vertical space display the count. Overall the spatial position is great for showing continous information (as noted in the lecture 4 data type slides). Vertical lines are displayed in red dotted lines to pop up against the solid blue of the histogram. These lines represent that different quartile values (25%, 50%, 75%) and can be used to noted the division of tiers.

```python
legendary=pkmn[(pkmn['Legendary']==1)]
nonlegendary=pkmn[(pkmn['Legendary']==0)] 
plt.scatter(legendary['Type 1'],legendary.Total,color='red',label='Legendary',marker="*",s=50) 
plt.scatter(nonlegendary['Type 1'],nonlegendary.Total,color='blue',label='Non - Legendary',s=25) 
plt.xlabel("Type 1")
plt.ylabel("Total Stats")
plt.legend()

plt.axhline(pkmn['Total'].mean(),linestyle='dashed',color='#88bbd6') 
plt.axhline(pkmn['Total'].quantile(0.75),linestyle='dashed',color='#88bbd6')
plt.axhline(pkmn['Total'].quantile(0.25),linestyle='dashed',color='#88bbd6')
plt.plot()

fig=plt.gcf()  #get the current figure using .gcf()
fig.set_size_inches(12,6) #set the size for the figure
plt.show()
```

### Visualization 2

In this visualization, we use a scatter plot to plot total stats vs the type 1 of the pokemon. The vertical space portray the distributions as well as min-max stat total for each type. There was an expectation that Legendary pokemon had higher stat total than non legendary pokemon. So they are displayed in a different color and symbol to emphasize their difference in status. Once again, lines are drawn represent that different quartile values (25%, 50%, 75%) and are displayed as a different color than the rest of the points.

### Other Visualization

Other visualization that was consider was a mult-histogram to show encode information about the types. However it was discarded because the many types forced the columns to be too narrow and hard to read. It is the scalability problem mentioned in Visualizing Analysis and Design (p. 150). Other visualization like boxplot can also be useful to portray quartile information. The difference would be using boxplots wouldn't be seeing the different bin distribution like in histograms. If we did boxplots by types, we could have also seen the quartile and distributions information by type.

### Conclusion

I like Visualization 2 better. This was because by using different colors and just simple point marks that save on space, we were able to encode more dimension than in visualization 1. Using the plot, we were also able to affirm that legendary pokemon has much better stats that others even in their type category.
